```python
from PIL import Image
from tqdm import tqdm
import glob

#We import the csv first
metadata = pd.read_csv("/content/drive/MyDrive/MAIS/ISIC2018_Task3_Training_LesionGroupings.csv")

#Now we import the training set images, let's create a function to import images from a folder so we can do this later
#We can transform each image as we import it, therefore saving some time
def images_from_folder(folderpath):
  image_list = []
  for filename in tqdm(glob.glob(folderpath+'/*.jpg')):
      im=Image.open(filename)
      image_list.append(im)
  return image_list

#Now call our function on our training set data (this might vary for each person using this colab file)
training = images_from_folder("/content/drive/MyDrive/MAIS/ISIC2018_Task3_Training_Input")
```

```python
#We now want to convert these images to np arrays
#First define a function to do this automatically for us
def image_to_array(imagename):
  img = image.img_to_array(imagename)  # convert image to numpy arry
  img = img.reshape((1,) + img.shape)  # reshape image
  return img/255.0 #to normalize

#Convert the training set
training_array = []
for pic in tqdm(training):
  training_array.append(image_to_array(pic))

#Print a test
print(training_array[0])
```
